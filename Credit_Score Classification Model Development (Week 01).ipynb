```python
outliers = ['Annual_Income', 'Monthly_Inhand_Salary', 'Delay_from_due_date', 'Changed_Credit_Limit', 'Num_Credit_Inquiries',
           'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Total_EMI_per_month', 'Amount_invested_monthly', 'Monthly_Balance']
```

```python
for column in outliers:
    df_credit.boxplot(column=column)
    plt.show()
```

```python
df_credit1 = df_credit.copy()
```

```python
df_credit1.head()
```

```python
#removing outliers
for column in outliers:
    Q1 = df_credit1[column].quantile(0.25)
    Q3 = df_credit1[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_credit1 = df_credit1[(df_credit1[column] >= lower_bound) & (df_credit1[column] <= upper_bound)]
```

```python
df_credit1.describe()
```

```python
df_credit1.nunique()
```

```python
#dropping some columns
df_credit_score = df_credit1.drop(['ID', 'Customer_ID', 'Month', 'Name', 'SSN', 'Occupation', 'Type_of_Loan', 'Payment_Behaviour', 'Payment_of_Min_Amount'], axis=1)
```

```python
df_credit_score.head()
```

```python
df_credit_score.shape
```

```python
#Scaling
from sklearn.preprocessing import MinMaxScaler
numerical_cols = df_credit_score.select_dtypes(include='number')
scalar = MinMaxScaler()
scaled_df_credit_score = scalar.fit_transform(numerical_cols)
scaled_df_credit_score = pd.DataFrame(scaled_df_credit_score, columns=numerical_cols.columns)
```

```python
df_credit_score[numerical_cols.columns] = scaled_df_credit_score
```

```python
df_credit_score.head()
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_credit_score['Credit_Score'] = le.fit_transform(df_credit_score['Credit_Score'])
df_credit_score['Credit_Mix'] = le.fit_transform(df_credit_score['Credit_Mix'])
```
